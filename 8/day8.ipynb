```python
f = open("input.txt", "r")
input = f.readline()
```

```python
print(len(input))
```

```python
width = 25
height = 6
resolution = width * height
numLayers = (int)((len(input))/resolution)
```

```python
layers = []
for i in range(numLayers):
    layers.append(input[i*resolution:(i*resolution+(resolution))])
print(layers)
```

```python
numZeroes = []
for layer in layers:
    numZeroes.append(layer.count('0'))
print(numZeroes)
layerWithLeastZeroes = numZeroes.index(min(numZeroes))
print("index of layer with least zeroes: "+str(layerWithLeastZeroes))
print("number of zeroes: "+str(numZeroes[layerWithLeastZeroes]))
numOnes = layers[layerWithLeastZeroes].count('1')
numTwos = layers[layerWithLeastZeroes].count('2')
print("Product of 1s and 2s: "+str(numOnes*numTwos))
```
