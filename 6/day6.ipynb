```python
with open("input.txt", "r") as f:
    inOrbitOfObject = {}
    for line in f:
        objects = line.split(')')
        objects[1] = objects[1][0:3]
        inOrbitOfObject[objects[1]] = objects[0]
```

```python
print(inOrbitOfObject)
```

```python
orbitersOf = {}
for key in inOrbitOfObject:
    if (inOrbitOfObject[key] not in orbitersOf):
        orbitersOf[inOrbitOfObject[key]] = []
    orbitersOf[inOrbitOfObject[key]].append(key)
```
