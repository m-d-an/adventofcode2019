```python
class IntCodeComputer:
    
    debug = False
    
    opTwo = [3,4,9]
    opThree = [5,6]
    opFour = [1,2,7,8]
    
    currentIndex = 0
    relativeBase = 0
    memory = None
            
    running = True
    lastOutput = -1
        
    def __init__(self):
        f = open("input.txt", "r")
        self.inputs = f.readline().split(',')
        self.inputs = list(map(int, self.inputs))
        #self.inputs = [109,1,204,-1,1001,100,1,100,1008,100,16,101,1006,101,0,99]
        #self.inputs = [1102,34915192,34915192,7,4,7,99,0]
        #self.inputs = [104,1125899906842624,99]
        f.close()       

        self.memory = self.inputs.copy()
        
    def compute(self):
        while (self.memory[self.currentIndex]!=99):    

            instruction = self.memory[self.currentIndex]
            opcode = 0
            params = []            
            address = []

            if (len(str(self.memory[self.currentIndex]))>2):
                # deconstruct instruction
                digits = list(map(int, str(instruction)))
                opcode = int(digits[-1])
                for i in range(len(digits)-2,0,-1):
                    params.append(int(digits[i-1]))
            else:
                opcode = instruction
                
            if (self.debug):
                print("IP: " + str(self.currentIndex) + 
                      ", Instruction: " + str(self.memory[self.currentIndex]) +
                     ", Opcode: " + str(opcode))
                print("Params: ")
                print(params)

            if (opcode in self.opFour):
                numIndices = 4
            elif (opcode in self.opThree):
                numIndices = 3
            elif (opcode in self.opTwo):
                numIndices = 2
            for i in range(numIndices-1):
                if (len(params)>i):
                    if(params[i]==1):
                        #immediate mode
                        address.append(self.currentIndex+i+1)
                    elif(params[i]==2):
                        #relative mode
                        address.append(self.memory[self.currentIndex+i+1]+self.relativeBase)
                    else:
                        #position mode
                        address.append(self.memory[self.currentIndex+i+1]) 
                else:
                    #position mode
                    address.append(self.memory[self.currentIndex+i+1]) 
                    
            # check if addresses exist and initialize with 0 otherwhise
            for ads in address:
                while (ads >= len(self.memory)):
                    self.memory.append(0)
                    
            self.currentIndex += numIndices
            
            if (self.debug):
                print("Adresses: ")    
                print(address) 

            if (opcode == 1):
                self.memory[address[2]] = self.memory[address[0]]+self.memory[address[1]]
            elif (opcode == 2):
                self.memory[address[2]] = self.memory[address[0]]*self.memory[address[1]]
            elif (opcode == 3):
                    number = int(input("Enter number: "))   
                    self.memory[address[0]] = number
            elif (opcode == 4):
                self.lastOutput = self.memory[address[0]]
                print(self.memory[address[0]])
                #return self.memory[address[0]]
            elif (opcode == 5):
                if (self.memory[address[0]] != 0):
                    self.currentIndex = self.memory[address[1]]
            elif (opcode == 6):
                if (self.memory[address[0]] == 0):
                    self.currentIndex = self.memory[address[1]]       
            elif (opcode == 7):
                if (self.memory[address[0]] < self.memory[address[1]]):
                    self.memory[address[2]] = 1
                else:
                    self.memory[address[2]] = 0
            elif (opcode == 8):
                if (self.memory[address[0]] == self.memory[address[1]]):
                    self.memory[address[2]] = 1
                else:
                    self.memory[address[2]] = 0
            elif (opcode == 9):
                self.relativeBase += self.memory[address[0]]
                
        if(self.debug):
            print("99 - Halt")
            
        self.running = False
        #print(self.lastOutput)
        #return self.lastOutput
```
