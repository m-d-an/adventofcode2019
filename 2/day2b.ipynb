```python
f = open("input.txt", "r")
```

```python
inputs = f.readline().split(',')
```

```python
inputs = list(map(int, inputs))
```

```python
print(inputs)
```

```python
memory = inputs.copy()

for i in range(0,99):
    for j in range(0,99):
        memory[1] = i
        memory[2] = j

        index = 0
        while (memory[index]!=99):
            try:
                if (memory[index] == 1):
                    memory[memory[index+3]] = memory[memory[index+1]]+memory[memory[index+2]]
                elif (memory[index] == 2):
                    memory[memory[index+3]] = memory[memory[index+1]]*memory[memory[index+2]]
                index += 4
                #print("Instruction Pointer: " + str(index) + ", Instruction: " + str(memory[index]))
            except IndexError:
                continue

        print(memory[0])
        if memory[0] == 19690720:
            break
        else:
            memory = inputs.copy()
    if memory[0] == 19690720:
            break
            
print("noun: "+str(memory[1]))
print("verb: "+str(memory[2]))
```
