```python
class Amplifier:
    
    opTwo = [3,4]
    opThree = [5,6]
    opFour = [1,2,7,8]
    
    currentIndex = 0
    memory = None
        
    phaseSetting = -1
    phaseSet = False
    
    running = True
    lastOutput = -1
        
    def __init__(self, phaseSetting):
        f = open("input.txt", "r")
        self.inputs = f.readline().split(',')
        self.inputs = list(map(int, self.inputs))
        #self.inputs = [3,26,1001,26,-4,26,3,27,1002,27,2,27,1,27,26,27,4,27,1001,28,-1,28,1005,28,6,99,0,0,5]
        #self.inputs = [3,52,1001,52,-5,52,3,53,1,52,56,54,1007,54,5,55,1005,55,26,1001,54,-5,54,1105,1,12,1,53,54,53,1008,54,0,55,1001,55,1,55,2,53,55,53,4,53,1001,56,-1,56,1005,56,6,99,0,0,0,0,10]
        f.close()       
        
        self.phaseSetting = phaseSetting
        self.memory = self.inputs.copy()
        
    def compute(self, inputSignal):
        while (self.memory[self.currentIndex]!=99):    

            instruction = self.memory[self.currentIndex]
            opcode = 0
            params = []            
            address = []

            if (len(str(self.memory[self.currentIndex]))>2):
                # deconstruct instruction
                digits = list(map(int, str(instruction)))
                opcode = int(digits[-1])
                for i in range(len(digits)-2,0,-1):
                    params.append(int(digits[i-1]))
            else:
                opcode = instruction

            if (opcode in self.opFour):
                numIndices = 4
            elif (opcode in self.opThree):
                numIndices = 3
            elif (opcode in self.opTwo):
                numIndices = 2
            for i in range(numIndices-1):
                if (len(params)>i and params[i]==1):
                    address.append(self.currentIndex+i+1)
                else:
                    address.append(self.memory[self.currentIndex+i+1]) 
            self.currentIndex += numIndices

            if (opcode == 1):
                self.memory[address[2]] = self.memory[address[0]]+self.memory[address[1]]
            elif (opcode == 2):
                self.memory[address[2]] = self.memory[address[0]]*self.memory[address[1]]
            elif (opcode == 3):
                if (not self.phaseSet):
                    self.memory[address[0]] = self.phaseSetting
                    self.phaseSet = True
                else:
                    self.memory[address[0]] = inputSignal
            elif (opcode == 4):
                self.lastOutput = self.memory[address[0]]
                return self.memory[address[0]]
                #print(memory[address[0]])
            elif (opcode == 5):
                if (self.memory[address[0]] != 0):
                    self.currentIndex = self.memory[address[1]]
            elif (opcode == 6):
                if (self.memory[address[0]] == 0):
                    self.currentIndex = self.memory[address[1]]       
            elif (opcode == 7):
                if (self.memory[address[0]] < self.memory[address[1]]):
                    self.memory[address[2]] = 1
                else:
                    self.memory[address[2]] = 0
            elif (opcode == 8):
                if (self.memory[address[0]] == self.memory[address[1]]):
                    self.memory[address[2]] = 1
                else:
                    self.memory[address[2]] = 0
        print("99 - Halt")
        self.running = False
        print(self.lastOutput)
        return self.lastOutput
```

```python
def runProgram(phaseSetting):
    amplifiers = []
    for i in range(5):
        amplifiers.append(Amplifier(phaseSetting[i]))

    outputSignal = 0
    running = True

    while running:
        for i in range(len(amplifiers)):
            outputSignal = amplifiers[i].compute(outputSignal)
            if (i==4 and amplifiers[i].running == False):
                running = False
                

    print(phaseSetting)
    print(outputSignal)
    return outputSignal
```

```python
phaseSetting = [0,0,0,0,0]
highestOutput = 0
for a in range(5,10):
    phaseSetting[0] = a
    for b in range(5,10):
        phaseSetting[1] = b
        for c in range(5,10):
            phaseSetting[2] = c
            for d in range(5,10):
                phaseSetting[3] = d
                for e in range(5,10):
                    phaseSetting[4] = e
                    if len(phaseSetting) == len(set(phaseSetting)):
                        output = runProgram(phaseSetting)
                        if (output > highestOutput):
                            highestOutput = output
                    else:
                        continue

print(highestOutput)                  
```
