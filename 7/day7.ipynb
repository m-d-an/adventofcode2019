```python
class Amplifier:
    
    opTwo = [3,4]
    opThree = [5,6]
    opFour = [1,2,7,8]
        
    def __init__(self):
        f = open("input.txt", "r")
        self.inputs = f.readline().split(',')
        self.inputs = list(map(int, self.inputs))
        #self.inputs = [3,15,3,16,1002,16,10,16,1,16,15,15,4,15,99,0,0]
        #self.inputs = [3,23,3,24,1002,24,10,24,1002,23,-1,23,101,5,23,23,1,24,23,23,4,23,99,0,0]
        #self.inputs = [3,31,3,32,1002,32,10,32,1001,31,-2,31,1007,31,0,33,1002,33,7,33,1,33,31,31,1,32,31,31,4,31,99,0,0,0]
        #print(self.inputs)
        f.close()       
        
    def compute(self, phaseSetting, inputSignal):
        phaseSet = False
        
        memory = self.inputs.copy()

        index = 0
        while (memory[index]!=99):    

            instruction = memory[index]
            opcode = 0
            params = []            
            address = []

            if (len(str(memory[index]))>2):
                # deconstruct instruction
                digits = list(map(int, str(instruction)))
                opcode = int(digits[-1])
                for i in range(len(digits)-2,0,-1):
                    params.append(int(digits[i-1]))
            else:
                opcode = instruction

            if (opcode in self.opFour):
                numIndices = 4
            elif (opcode in self.opThree):
                numIndices = 3
            elif (opcode in self.opTwo):
                numIndices = 2
            for i in range(numIndices-1):
                if (len(params)>i and params[i]==1):
                    address.append(index+i+1)
                else:
                    address.append(memory[index+i+1]) 
            index += numIndices

            if (opcode == 1):
                memory[address[2]] = memory[address[0]]+memory[address[1]]
            elif (opcode == 2):
                memory[address[2]] = memory[address[0]]*memory[address[1]]
            elif (opcode == 3):
                if (not phaseSet):
                    memory[address[0]] = phaseSetting
                    phaseSet = True
                else:
                    memory[address[0]] = inputSignal
            elif (opcode == 4):
                return memory[address[0]]
                #print(memory[address[0]])
            elif (opcode == 5):
                if (memory[address[0]] != 0):
                    index = memory[address[1]]
            elif (opcode == 6):
                if (memory[address[0]] == 0):
                    index = memory[address[1]]       
            elif (opcode == 7):
                if (memory[address[0]] < memory[address[1]]):
                    memory[address[2]] = 1
                else:
                    memory[address[2]] = 0
            elif (opcode == 8):
                if (memory[address[0]] == memory[address[1]]):
                    memory[address[2]] = 1
                else:
                    memory[address[2]] = 0
```

```python
def runProgram(phaseSetting):
    amplifiers = []
    for i in range(5):
        amplifiers.append(Amplifier())

    outputSignal = 0

    for i in range(len(amplifiers)):
        outputSignal = amplifiers[i].compute(phaseSetting[i], outputSignal)

    print(phaseSetting)
    print(outputSignal)
    return outputSignal
```

```python
phaseSetting = [0,0,0,0,0]
highestOutput = 0
for a in range(5):
    phaseSetting[0] = a
    for b in range(5):
        phaseSetting[1] = b
        for c in range(5):
            phaseSetting[2] = c
            for d in range(5):
                phaseSetting[3] = d
                for e in range(5):
                    phaseSetting[4] = e
                    if len(phaseSetting) == len(set(phaseSetting)):
                        output = runProgram(phaseSetting)
                        if (output > highestOutput):
                            highestOutput = output
                    else:
                        continue

print(highestOutput)                  
```
